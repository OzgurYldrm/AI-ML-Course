```python
import matplotlib.pyplot as plt
```

# Temel Bileşenler

**title** -> Grafik başlığı     
**xlabel** -> X ekseni ismi     
**ylabel** -> y ekseni ismi     
**label** -> Grafiğin detayını belirtebilir.Çoklu grafiklerde daha işlevsel olabilir.        
**legend()** -> label parametresinin görünür olmasını sağlar.       
**grid()** -> Yardımcı çizgilerin etkinleştirilmesini sağlar.       
**xticks/yticks** -> Eksenlerde gösterilecek değerleri belirtebilirsiniz.     
**xlim/ylim** -> Eksenlerin değer aralığını belirtebilirsiniz.

**color** -> çizginin rengini belirler. (red-r , blue-b , HEX kodları, (0.2,0.1,0.2) RGB kodları)       
**linestyle** -> Çizgi biçimini belirler. ("-", "--" , "-.", ":")       
**linewidth** -> Çizgi kalınlığını belirler.        
**marker** -> Veri noktalarını işaretlemek için sembol ekler. (o, s, ^ , *, ., +)      
**markeredgecolor** -> Marker'ın kenar rengini belirler     

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, label="Doğrusal İlişki",marker="*",markeredgecolor="black",color="r")

plt.title("Basit Grafik")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.legend(loc="upper left")

plt.grid(color='green', linestyle='--', linewidth=0.5)

#plt.xticks([0, 2, 4])
#plt.yticks([0, 2, 4])

#plt.xlim(0, 50)
#plt.ylim(0, 50)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]   # 1. veri
y2 = [1, 3, 5, 7, 9]    # 2. veri

plt.plot(x, y1, label="Doğrusal Artış", color="blue", marker="o")
plt.plot(x, y2, label="Diğer Seri", color="red", marker="^")

plt.title("Tek Grafikte İki Plot")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.legend()
plt.grid(True)

plt.show()
```

**figure** -> grafiğin tuvalidir. Doğrudan oluşturulmadan sadece plot kullanılırsa matplotlib otomatik oluşturur.

```python
plt.figure(figsize=(18,15), dpi=100)

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color="blue", marker="o")
plt.title("Figure Örneği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

plt.show()
```

# Grafik Türleri

**çizgi grafiği**

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8,15]

plt.plot(x, y, label="Doğrusal", color="blue", marker="o")
plt.title("Çizgi Grafiği (Line Plot)")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.legend()
plt.show()
```

**bar grafiği**

```python
x = ["A", "B", "C", "D"]
y = [10, 20, 15, 25]

plt.bar(x, y, color="green")
plt.title("Bar Grafiği (Dikey)")
plt.xlabel("Kategori")
plt.ylabel("Değer")
plt.show()
```

```python
x = ["A", "B", "C", "D"]
y = [10, 20, 15, 25]

plt.barh(x, y, color="orange")
plt.title("Bar Grafiği (Yatay)")
plt.xlabel("Değer")
plt.ylabel("Kategori")
plt.show()
```

**scatter plot**

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y, color="red", marker="^")
plt.title("Scatter Grafiği (Dağılım)")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.show()
```

**histogram grafiği**

```python
data = [7, 8, 5, 6, 4, 5, 5, 6, 7, 8, 8, 9, 4, 5, 6, 7]

plt.hist(data, bins=5, color="purple", edgecolor="black")
plt.title("Histogram")
plt.xlabel("Değerler")
plt.ylabel("Frekans")
plt.show()
```

```python
labels = ["A", "B", "C", "D"]
values = [15, 30, 45, 10]

plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart (Pasta Grafiği)")
plt.show()
```

# Subplots

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 9, 16, 25]

fig, axs = plt.subplots(1, 2, figsize=(20,8))

axs[0].plot(x, y1, color="blue", marker="o")
axs[0].set_title("Doğrusal Grafik")

axs[1].plot(x, y2, color="red", marker="^")
axs[1].set_title("Kare Fonksiyonu")

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 9, 16, 25]
y3 = [2, 3, 5, 7, 11]
y4 = [5, 7, 6, 8, 7]

fig, axs = plt.subplots(2, 2, figsize=(8, 6))  # 2 satır, 2 sütun

# 1. grafik
axs[0, 0].plot(x, y1, color="blue")
axs[0, 0].set_title("Doğrusal")

# 2. grafik
axs[0, 1].plot(x, y2, color="red")
axs[0, 1].set_title("Kare")

# 3. grafik
axs[1, 0].bar(x, y3, color="green")
axs[1, 0].set_title("Bar Grafiği")

# 4. grafik
axs[1, 1].scatter(x, y4, color="purple")
axs[1, 1].set_title("Scatter")

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig, axs = plt.subplots(2, 2, figsize=(8, 6))  # 2x2 grid
axs = axs.flatten()  # 1D array haline getir

axs[0].plot(x, y, color="blue")
axs[0].set_title("Grafik 1")

axs[1].bar(x, y, color="green")
axs[1].set_title("Grafik 2")

axs[2].scatter(x, y, color="red")
axs[2].set_title("Grafik 3")

axs[3].hist(y, bins=5, color="purple")
axs[3].set_title("Grafik 4")

plt.tight_layout()
plt.show()
```
