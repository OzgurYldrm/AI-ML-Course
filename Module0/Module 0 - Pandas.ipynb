# groupby - aggregation

**groupby** -> Belirli sütunlar göz önünde bulundurularak gruplama yapılmasını ifade eder. Sütun içerisinde x,y,z olmak üzere 3 farklı kategorik değer varsa, gruplama yapıldıktan sonra her satır kendi ait olduğu gruba göre değerlendirilir. pandas groupby objesi normal df gibi davranmaz.

```python
df.head(5)
```

```python
df.groupby("embark_town").first()
```

Gruplama yapıldıktan sonra her grup için mean, median ... değerleri hesaplanabilir. 

```python
df.groupby("embark_town")["age"].mean()
```

gruplama yapıldıktan sonra aggregation ile diğer sütunlarla ilgili özellikler hesaplanabilir.

```python
df.groupby("embark_town").agg(
    grouped_mean_age = ("age","mean"),
    grouped_median_age = ("age","median"),
)
```

Gruplama yapıldıktan sonra transform ile yapılan işlemler tekrardan bütün satırlara geri gönderilebilir. (Ekleme: lambda fonksiyonu kullanımı zorunlu değildir. Normal fonksiyonlar da kullanılabilir.)

```python
df["age_trasnformed"] = df.groupby("embark_town")["age"].transform("mean")
```

```python
df["normalize"] = df.groupby("embark_town")["age"].transform(lambda x: x / x.max())
```

# concat-merge

**merge** sql join mantığı ile birleştirmek için kullanılır.

```python
df_heroes = pd.DataFrame({
    "isim": ["Batman", "Superman", "Iron Man", "Spider-Man"],
    "takım": ["Justice League", "Justice League", "Avengers", "Avengers"]
})

df_scores = pd.DataFrame({
    "isim": ["Batman", "Superman", "Iron Man", "Spider-Man", "Hulk"],
    "puan": [90, 85, 88, 92, 95]
})
```

```python
inner = pd.merge(df_heroes, df_scores, on="isim", how="inner")
inner
```

```python
left = pd.merge(df_heroes, df_scores, on="isim", how="left")
left
```

```python
outer = pd.merge(df_heroes, df_scores, on="isim", how="outer")
outer
```

# datetime

**datetime** pandas içerisinde tarih verileri ile işlem yapmak için kullanılır.

```python
df = pd.DataFrame({
    "isim": ["Batman", "Iron Man", "Spider-Man"],
    "doğum": ["1939-05-01", "1963-03-01", "1962-08-01"]
})

df["doğum"] = pd.to_datetime(df["doğum"])
print(df.dtypes)
```

```python
df
```

datetime sütunundan belirli özellikler çıkartılabilir.

```python
df["yıl"] = df["doğum"].dt.year
df["ay"] = df["doğum"].dt.month
df["gün"] = df["doğum"].dt.day
```

date ile aritmetik işlemler yapılabilir.

```python
bugün = pd.Timestamp("2025-08-20")
df["yaş_gün"] = (bugün - df["doğum"]).dt.days
df["yaş_yıl"] = df["yaş_gün"] // 365

print(df[["isim", "yaş_yıl"]])
```

date kullanılarak filtreleme yapılabilir.

```python
filtre = df[df["doğum"] > "1960-01-01"]
print(filtre)
```

**date_range** ile istenen formatta belirli bir date aralığı oluşturulabilir.

```python
tarih_aralığı = pd.date_range(start="2025-01-01", end="2025-01-10", freq="D")
print(tarih_aralığı)
```
