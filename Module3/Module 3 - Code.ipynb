# Handling Outliers

```python
column = "LotFrontage"
```

```python
data = df['LotFrontage']
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('Boxplot')
plt.xlabel('Değerler')
plt.show()
```

```python
Q1 = df[column].quantile(0.25)
Q3 = df[column].quantile(0.75)
IQR = Q3 - Q1
df_outlier_deleted = df[df[column].between(Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)]

data = df_outlier_deleted['LotFrontage']
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('Boxplot')
plt.xlabel('Değerler')
plt.show()
```

```python
Q1 = df[column].quantile(0.25)
Q3 = df[column].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df_outlier_winsorized = df[column].clip(lower, upper)

data = df_outlier_winsorized
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('Boxplot')
plt.xlabel('Değerler')
plt.show()
```

# Missing Values

```python
missing_counts = df.isnull().sum()
missing_percent = (missing_counts / len(df)) * 100
dtypes = df.dtypes
missing_df = pd.DataFrame({
    'Missing Count': missing_counts,
    'Missing Percentage (%)': missing_percent,
    'Dtype': dtypes
})
missing_df = missing_df[missing_df['Missing Count'] > 0]
missing_df = missing_df.sort_values(by='Missing Percentage (%)', ascending=False)
missing_df
```

```python
threshold = df.shape[1] * (0.1)
df[df.isnull().sum(axis=1) > threshold]
```

```python
df.shape
```

```python
drop_columns = missing_df[missing_df["Missing Percentage (%)"] > 40].index.to_list()
df.drop(columns=drop_columns, inplace=True)
df.shape
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
```

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].mean())
#df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
df["MasVnrArea"] = imputer.fit_transform(pd.DataFrame(df["MasVnrArea"]))
```

```python
df['GarageCond'] = df['GarageCond'].fillna("Unknown")
```

```python
cat_cols = df.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode(dropna=True)[0])
```

```python
df_group = pd.DataFrame({
    'meslek': ['Mühendis', 'Doktor', 'Mühendis', 'Öğretmen', 'Doktor', 'Öğretmen'],
    'maaş': [np.nan,10000, 11000, 12000,np.nan, np.nan]
})
print(df_group)
df_group['maaş'] = df_group.groupby('meslek')['maaş'].transform(
    lambda x: x.fillna(x.mean())
)
print("*"*25)
print(df_group)
```

```python
data = np.random.normal(loc=50, scale=10, size=(1000, 1))
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')
plt.title('Orijinal Veri')
plt.xlabel('Değer')
plt.ylabel('Yoğunluk')

plt.subplot(1, 2, 2)
plt.hist(data_scaled, bins=30, density=True, alpha=0.6, color='green')
plt.title('StandardScaler ile Ölçeklenmiş Veri')
plt.xlabel('Değer')
plt.ylabel('Yoğunluk')

plt.tight_layout()
plt.show()

print(f"Orijinal veri ortalaması: {np.mean(data):.4f}, std: {np.std(data):.4f}")
print(f"Ölçeklenmiş veri ortalaması: {np.mean(data_scaled):.4f}, std: {np.std(data_scaled):.4f}")
```

```python
data = np.random.exponential(scale=2, size=(1000, 1))
scaler = RobustScaler()
data_scaled = scaler.fit_transform(data)

plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')
plt.title('Orijinal Veri')
plt.xlabel('Değer')
plt.ylabel('Yoğunluk')

plt.subplot(1, 2, 2)
plt.hist(data_scaled, bins=30, density=True, alpha=0.6, color='green')
plt.title('Robust Scaler ile Ölçeklenmiş Veri')
plt.xlabel('Değer')
plt.ylabel('Yoğunluk')

plt.tight_layout()
plt.show()

print(f"Orijinal veri ortalaması: {np.mean(data):.4f}, std: {np.std(data):.4f}")
print(f"Ölçeklenmiş veri ortalaması: {np.mean(data_scaled):.4f}, std: {np.std(data_scaled):.4f}")
```

# Feature Selection

```python
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(20,15))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Map of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_regression

df_numeric = df.select_dtypes(include='number')
X = df_numeric.drop('SalePrice', axis=1)
y = df_numeric['SalePrice']
mi_scores = mutual_info_regression(X, y)
mi_df = pd.DataFrame({'Feature': X.columns, 'Mutual Information': mi_scores})
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)

plt.figure(figsize=(14,8))
sns.barplot(data=mi_df, x='Mutual Information', y='Feature', palette='mako')
plt.title('Mutual Information (Numeric Features Only)')
plt.tight_layout()
plt.show()
```
